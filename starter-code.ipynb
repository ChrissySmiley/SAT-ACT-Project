##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
#original standard deviation function, works
def std_dev(df,column):
    '''
    df: a pandas df
    column: a string, one of the columns names in df
    '''
    cnt_n = len(df[column])
    mean_ = sum(df[column])/cnt_n
    sum_of_squares = 0
    for val in df[column]:
        sum_of_squares += (val - mean_)**2
    variance = sum_of_squares/cnt_n
    std_dev = variance ** .5
    return std_dev
```

##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

H(0) - that they're the same
H(a) - that they're different

```python
from scipy import stats
sat_participation = scores['SAT_Participation']
act_participation = scores['ACT_Participation']
stats.ttest_ind(sat_participation, act_participation)
```

Since the P value is less than 5, we reject the null hypothesis and the result is statistically significant.
